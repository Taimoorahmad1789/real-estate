```python
def price_to_crore(price_str):
    """
    Convert a price string like '5,666,150 PKR' into CRORE format.
    Returns a float value in crore.
    """
    import re
    
    # Remove PKR and commas
    clean = re.sub(r'[^\d.]', '', price_str)
    
    if clean == "":
        return None
    
    price = float(clean)
    
    # Convert to crore
    crore_value = price / 10000000  # 1 crore = 10,000,000
    
    return round(crore_value, 2)  # Rounded to 2 decimals

    df['price_in_crore'] = df['price'].apply(price_to_crore)
```

```python
def convert_price_to_crore(price_str):
    """
    Converts a price value like '5,666,150 PKR' to crore format.
    Example: '5,666,150 PKR' -> 0.57 crore
    """
    import re
    
    # Remove PKR, commas, and spaces → keep only digits
    numeric = re.sub(r'[^\d]', '', price_str)
    
    if numeric == "":
        return None
    
    price = float(numeric)
    
    # Convert to crore (1 crore = 10,000,000)
    crore_value = price / 10000000
    
    return round(crore_value, 2)

    df['price_in_crore'] = df['price'].apply(price_to_crore)
```

```python
df['price_in_crore'] = df['price'].apply(convert_price_to_crore)
```
