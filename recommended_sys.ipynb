```python
def convert_to_crore(price_string):
    """Extract PKR amount and convert to Crore"""
    # Extract numeric values from the string
    numbers = re.findall(r'[\d,]+', price_string)
    
    if not numbers:
        return price_string
    
    # Get the first number (price in PKR)
    price_pkr = int(numbers[0].replace(',', ''))
    
    # Convert PKR to Crore (1 Crore = 10,000,000 PKR)
    price_crore = price_pkr / 10000000
    
    # Replace PKR amount with Crore amount
    updated_string = price_string.replace(numbers[0], f"{price_crore:.2f} Crore")
    
    return updated_string

# Apply the conversion function to PriceDetails column
df['PriceDetails'] = df['PriceDetails'].apply(convert_to_crore)
```

```python
def convert_facilities_to_string(facilities):
    """Convert list format string to clean string"""
    # Remove brackets [ and ]
    facilities = facilities.replace('[', '').replace(']', '')
    
    # Remove quotes if any
    facilities = facilities.replace("'", "").replace('"', '')
    
    return facilities

# Apply conversion to TopFacilities column
df['FacilitiesStr'] = df['TopFacilities'].apply(convert_facilities_to_string)
```

```python
def recommend_properties(property_name, cosine_sim=cosine_sim1):
    # Get the index of the property that matches the name
    idx = df.index[df['PropertyName'] == property_name].tolist()[0]

    # Get the pairwise similarity scores with that property
    sim_scores = list(enumerate(cosine_sim1[idx]))

    # Sort the properties based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar properties
    sim_scores = sim_scores[1:6]

    # Get the property indices
    property_indices = [i[0] for i in sim_scores]
    
    recommendations_df = pd.DataFrame({
        'PropertyName': df['PropertyName'].iloc[property_indices],
        'SimilarityScore': sim_scores
    })

    # Return the top 10 most similar properties
    return recommendations_df
```

```python
import pandas as pd
import json
import re

# Load the dataset
df_apartments = pd.read_excel('Islamabad_Apartments_250_Final.xlsx')

# Function to parse PriceDetails and extract features
def parse_price_details(detail_str):
    """
    Parse price details string and extract building type, area, and price
    Format: "4.50 Crore | 1000 sqft | Monthly Rent: PKR 45,000"
    """
    try:
        extracted = {}
        
        # Extract price in Crore
        price_match = re.search(r'([\d.]+)\s*Crore', detail_str)
        if price_match:
            price = float(price_match.group(1))
            extracted['price'] = price
        
        # Extract area in sqft
        area_match = re.search(r'(\d+)\s*sqft', detail_str)
        if area_match:
            area = float(area_match.group(1))
            extracted['area'] = area
        
        return extracted
    except:
        return {}

# Function to get building type from PropertySubName
def get_building_type(subname):
    """Extract BHK type and building type from PropertySubName"""
    bhk_match = re.search(r'(\d+)\s*BHK', subname)
    if bhk_match:
        bhk = f"{bhk_match.group(1)} BHK"
    else:
        bhk = None
    
    building_type = None
    if 'Apartment' in subname:
        building_type = 'Apartment'
    elif 'Villa' in subname:
        building_type = 'Villa'
    elif 'Floor' in subname:
        building_type = 'Independent Floor'
    
    return bhk, building_type

# Extract location from PropertyName (sirf pehla part)
def extract_location(property_name):
    """Extract location name - sirf pehla part dashes sai pehle"""
    return property_name.split(' - ')[0].strip()

# Transform the dataset
data_refined = []

for idx, row in df_apartments.iterrows():
    # Extract location name (sirf location, baki details remove)
    location = extract_location(row['PropertyName'])
    
    # Extract BHK type and building type
    bhk, building_type = get_building_type(row['PropertySubName'])
    
    # Parse price details
    price_data = parse_price_details(row['PriceDetails'])
    
    # Create base row with location instead of full PropertyName
    new_row = {
        'PropertyName': location,  # Sirf location name - column name same
        'PropertySubName': row['PropertySubName'],
        'LocationAdvantages': row['LocationAdvantages'],
        'TopFacilities': row['TopFacilities'],
        'Link': row['Link']
    }
    
    # Add building type, area, and price for all BHK configurations
    for config in ['1 BHK', '2 BHK', '3 BHK', '4 BHK', '5 BHK', '6 BHK', '1 RK', 'Land']:
        # Set building type only for matching BHK
        if bhk == config:
            new_row[f'building_type_{config}'] = building_type
            new_row[f'area_low_{config}'] = price_data.get('area')
            new_row[f'area_high_{config}'] = price_data.get('area')
            new_row[f'price_low_{config}'] = price_data.get('price')
            new_row[f'price_high_{config}'] = price_data.get('price')
        else:
            new_row[f'building_type_{config}'] = None
            new_row[f'area_low_{config}'] = None
            new_row[f'area_high_{config}'] = None
            new_row[f'price_low_{config}'] = None
            new_row[f'price_high_{config}'] = None
    
    data_refined.append(new_row)

# Create the final refined dataframe
df_final_refined = pd.DataFrame(data_refined)

# Set Location as index
df_final_refined_indexed = df_final_refined.set_index('PropertyName')

# Fill empty building type columns with location name
print("--- Filling empty building type columns with location names ---")
building_cols = [col for col in df_final_refined.columns if 'building_type_' in col]
for col in building_cols:
    building_name = col.replace('building_type_', '')
    # Fillna aur empty strings ko location name se replace karo
    df_final_refined_indexed[col] = df_final_refined_indexed[col].fillna(building_name).replace('', building_name)
    print(f"✓ {col} -> filled with location-based building type")

# Display sample
print("\n✓ Dataset Transformation Complete!\n")
print("Sample data (first 5 rows):")
print(df_final_refined_indexed.head())
print(f"\n✓ Total rows: {len(df_final_refined_indexed)}")
print(f"✓ Total columns: {len(df_final_refined_indexed.columns)}")

# Save to Excel
df_final_refined_indexed.to_excel('Islamabad_Apartments_Transformed_Indexed.xlsx')
print("\n✓ File saved as 'Islamabad_Apartments_Transformed_Indexed.xlsx' (Location as index)")

# Save to CSV
df_final_refined.to_csv('Islamabad_Apartments_Transformed.csv', index=False)
print("✓ File saved as 'Islamabad_Apartments_Transformed.csv'")

# Display column breakdown
print("\n--- Column Structure ---")
print("Base columns: PropertyName, PropertySubName, LocationAdvantages, TopFacilities, Link")
print("\nFor each BHK type (1 BHK, 2 BHK, 3 BHK, 4 BHK, 5 BHK, 6 BHK, 1 RK, Land):")
print("  - building_type_{BHK}")
print("  - area_low_{BHK}")
print("  - area_high_{BHK}")
print("  - price_low_{BHK} (in Crore)")
print("  - price_high_{BHK} (in Crore)")

df_final_refined_v2 = df_final_refined_indexed
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd

def recommend_properties_with_scores(property_name, top_n=5):
    """
    Recommend properties based on cosine similarity
    """
    try:
        # Check if property exists
        if property_name not in ohe_df_normalized.index:
            print(f"✗ Property '{property_name}' not found!")
            return None
        
        print(f"\n✓ Found property: '{property_name}'")
        
        # Calculate cosine similarity
        print("✓ Calculating cosine similarity...")
        cosine_sim = cosine_similarity(ohe_df_normalized)
        
        # Ensure it's a numpy array
        cosine_sim = np.array(cosine_sim)
        cosine_sim = np.nan_to_num(cosine_sim, nan=0.0)
        
        print(f"✓ Cosine similarity shape: {cosine_sim.shape}")
        
        # FIX: Use where() instead of get_loc() to get proper index
        property_indices = np.where(ohe_df_normalized.index == property_name)[0]
        
        if len(property_indices) == 0:
            print(f"✗ Property '{property_name}' not found!")
            return None
        
        property_idx = property_indices[0]  # Get first match
        print(f"✓ Property index: {property_idx}")
        
        # Get similarity scores for this property
        sim_scores = cosine_sim[property_idx]
        print(f"✓ Similarity scores type: {type(sim_scores)}, shape: {sim_scores.shape}")
        
        # Create list of (index, score) tuples
        sim_list = [(i, float(sim_scores[i])) for i in range(len(sim_scores))]
        
        # Sort by score (descending)
        sorted_scores = sorted(sim_list, key=lambda x: x[1], reverse=True)
        
        # Get top_n (skip first - it's the property itself)
        top_indices = [i[0] for i in sorted_scores[1:top_n+1]]
        top_scores = [i[1] for i in sorted_scores[1:top_n+1]]
        
        # Get property names
        recommendations = ohe_df_normalized.index[top_indices].tolist()
        
        # Create dataframe
        recommendations_df = pd.DataFrame({
            'PropertyName': recommendations,
            'Similarity_Score': top_scores
        })
        
        print(f"\n✓ Top {top_n} recommendations for '{property_name}':\n")
        print(recommendations_df.to_string(index=False))
        
        return recommendations_df
    
    except Exception as e:
        print(f"✗ Error: {str(e)}")
        import traceback
        traceback.print_exc()
        return None

# Test the function
print("=" * 60)
print("TESTING RECOMMENDATION FUNCTION")
print("=" * 60)
result = recommend_properties_with_scores('DHA Phase 1')
```

```python
def distance_to_meters(distance_str):
    try:
        if 'Km' in distance_str or 'KM' in distance_str:
            return float(distance_str.split()[0]) * 1000
        elif 'Meter' in distance_str or 'meter' in distance_str:
            return float(distance_str.split()[0])
        else:
            return None
    except:
        return None
```

```python
import pandas as pd
import numpy as np

# Load your apartments dataset
df = pd.read_excel('Islamabad_Apartments_250_Final.xlsx')

# Define major landmarks/locations (exact column names like your example)
landmarks = [
    'Jinnah Avenue', 'Blue Area', 'Centaurus Mall', 'Serena Hotel', 'PIMS Hospital',
    'Aabpara Market', 'Atrium Mall', 'Rawal Lake', 'Pakistan Monument', 'Convention Center',
    'Rose & Jasmine Road', 'GT Road', 'Expressway', 'Airport Road', 'Margalla Hills',
    'Shakarparian', 'Civic Center', 'Melody Market', 'Hyperstar', 'F-9 Park'
]

# Create location_matrix dictionary
location_matrix = {}

for index, row in df.iterrows():
    distances = {}
    
    # For each landmark, randomly assign distance or NaN
    for landmark in landmarks:
        # 50% chance of having a distance, 50% NaN
        if np.random.random() > 0.5:
            # Random distance between 500m and 25000m (0.5 to 25 KM)
            distance = np.random.randint(500, 25000)
            distances[landmark] = distance
        else:
            distances[landmark] = np.nan
    
    location_matrix[index] = distances

# Convert dictionary to dataframe (index = property ID)
location_df = pd.DataFrame.from_dict(location_matrix, orient='index')

# Display info
print("Location Matrix Created!")
print(f"Shape: {location_df.shape} (rows × columns)")
print(f"Total properties: {len(location_df)}")
print(f"Total landmarks: {len(location_df.columns)}")
print("\nFirst few rows:")
print(location_df.head(10))




# Display sample
print("\n--- Sample Output (like your screenshot) ---")
location_df.head()
```
