```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'] = df['society'].str.replace('nan','independent')
```

```python
def convert_price_to_crore(price_str):
    import re
    
    # Ensure it's a string
    price_str = str(price_str)
    
    # Extract only digits
    numeric = re.sub(r'[^\d]', '', price_str)
    
    if not numeric:
        return None
    
    value = float(numeric)
    return round(value / 10000000, 2)
df['price_in_crore'] = df['price'].apply(convert_price_to_crore)
```

```python
df.head()
```

```python
df.drop(columns=['price'], inplace=True)
```

```python
#df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

```python
cols = list(df.columns)

# current index of price_per_sqft
pps_index = cols.index('price_per_sqft')

# remove both columns first
cols.remove('price_per_sqft')
cols.remove('areaWithType')

# insert areaWithType where price_per_sqft was originally
cols.insert(pps_index, 'areaWithType')

# insert price_per_sqft at 2nd last position
cols.insert(len(cols), 'price_per_sqft')   # this puts it last
cols.insert(len(cols)-1, cols.pop())       # this moves it to 2nd last

# apply new column order
df = df[cols]
```
